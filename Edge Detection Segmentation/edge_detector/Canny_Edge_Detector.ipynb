```python
%matplotlib inline
from utils import utils
import canny_edge_detector as ced
import math
import numpy as np
```

```python
imgs = utils.load_data()
utils.visualize(imgs, 'gray')
```

```python
detector = ced.cannyEdgeDetector(imgs, sigma=1.4, kernel_size=5, lowthreshold=0.09, highthreshold=0.17, weak_pixel=100)
```

```python
imgs_final = detector.detect()
```

```python
utils.visualize(imgs_final, 'gray')
```

```python
def rmse(r, t):
    #r is the original image, t is the segmented image
    score = np.sqrt((np.sum(r**2)/np.sum((r-t)**2))/(t.shape[0]*t.shape[1]*3))
    return score
```

```python
def psnr(r, t):
    #r is the original image, t is the segmented image
    score = 10*math.log10((r.max())**2/(np.sqrt((np.sum(r**2)/np.sum((r-t)**2))/(t.shape[0]*t.shape[1]*3))))
    return score
```

```python
print("rmse:")
for i in range (10):
    print(rmse(imgs[i], imgs_final[i]))

print("psnr:")
for i in range (10):
    print(psnr(imgs[i], imgs_final[i]))
```
